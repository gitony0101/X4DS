```julia
using CSV
using DataFrames
using Plots
using StatsPlots

using DecisionTree
```

```julia
hitData = CSV.File("dataset/Hitters.csv") |> DataFrame;
```

```julia
# Figure 8.2
scatter(hitData.Years, hitData.Hits, xlabel="Years", ylabel="Hits")
vline!([4.5], lw=2)
plot!([4.5, 25], [117.5, 117.5], lw=2)
```
