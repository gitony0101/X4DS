```julia
using CSV
using DataFrames
using Plots
using StatsPlots
```

## Advertising Data

```julia
adsData = CSV.File("dataset/Advertising.csv") |> DataFrame;
```

```julia
sales = adsData.sales
tv = adsData.TV
radio = adsData.radio
newspaper = adsData.newspaper

p1 = scatter(tv, sales, xlabel="TV", ylabel="Sales")
p2 = scatter(radio, sales, xlabel="Radio")
p3 = scatter(newspaper, sales, xlabel="Newspaper")

plot(p1, p2, p3, layout=(1, 3), legend=false, markercolor="red")
```

## Income Data

```julia
incomeData = CSV.File("dataset/Income1.csv") |> DataFrame;
```

```julia
edu = incomeData.Education
income = incomeData.Income

scatter(edu, income, legend=false, xlabel="Years of Education", ylabel="Income", markercolor="red")
```
