```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
# URL of the webpage you want to scrape
url = 'https://www.summersidetoyota.com/en/new-inventory'
```

```python
# Initialize the Chrome WebDriver
driver = webdriver.Chrome()

# Open the webpage
driver.get(url)

# Wait for the dynamic content to load
wait = WebDriverWait(driver, 10)

# Find the div elements with the specified class
specific_divs = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'listing-new-tile-wrapper.false')))

# Extract data from the elements
div_texts = [div.text for div in specific_divs]


# Close the WebDriver
driver.quit()
```

```python
div_texts
```

```python
text_data = [div_texts[_].split('\n') for _ in range(len(div_texts
))]
```

```python
text_data 
```

```python
# Extracting required information
extracted_data = []
for sublist in text_data:
    price = None
    vin = None
    for item in sublist:
        if item.startswith('$'):
            price = item
        elif item.startswith('VIN'):
            vin = item.split()[-1]
    extracted_data.append([sublist[0], sublist[1], price, vin])

# Converting to DataFrame
df = pd.DataFrame(extracted_data, columns=['Model', 'Drive', 'Price', 'VIN'])
```

```python
# Initialize the Chrome WebDriver
driver = webdriver.Chrome()

# Open the webpage
driver.get(url)
```

```python
# Find the specific div elements containing vehicle information
vehicle_info_divs = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'vehicle-card')))

# Initialize empty lists to store data
year_make_model_list = []
condition_list = []
body_style_list = []
odometer_list = []
stock_number_list = []
vin_list = []
color_list = []
msrp_list = []
internet_price_list = []
trim_list = []
transmission_list = []
class_list = []

# Loop through each vehicle info div and extract information
for vehicle_info_div in vehicle_info_divs:
    # Extracting individual vehicle information
    year_make_model = vehicle_info_div.find_element(By.CLASS_NAME, 'vehicle-card__title').text
    condition = vehicle_info_div.get_attribute('data-vehicle-condition')
    body_style = vehicle_info_div.get_attribute('data-vehicle-body-style')
    odometer = vehicle_info_div.get_attribute('data-vehicle-odo')
    stock_number = vehicle_info_div.get_attribute('data-vehicle-stock')
    vin = vehicle_info_div.get_attribute('data-vehicle-vin')
    color = vehicle_info_div.get_attribute('data-vehicle-colour')
    msrp = vehicle_info_div.get_attribute('data-vehicle-msrp')
    internet_price = vehicle_info_div.get_attribute('data-vehicle-internet-price')
    trim = vehicle_info_div.get_attribute('data-vehicle-trim')
    transmission = vehicle_info_div.get_attribute('data-vehicle-transmission')
    vehicle_class = vehicle_info_div.get_attribute('data-vehicle-class')
    
    # Split the year, make, and model from the title
    year_make_model_parts = year_make_model.split(' ', 1)
    year_make_model_list.append(year_make_model)
    condition_list.append(condition)
    body_style_list.append(body_style)
    odometer_list.append(odometer)
    stock_number_list.append(stock_number)
    vin_list.append(vin)
    color_list.append(color)
    msrp_list.append(msrp)
    internet_price_list.append(internet_price)
    trim_list.append(trim)
    transmission_list.append(transmission)
    class_list.append(vehicle_class)

# Create a DataFrame from the lists
df = pd.DataFrame({
    'Year_Make_Model': year_make_model_list,
    'Condition': condition_list,
    'Body_Style': body_style_list,
    'Odometer': odometer_list,
    'Stock_Number': stock_number_list,
    'VIN': vin_list,
    'Color': color_list,
    'MSRP': msrp_list,
    'Internet_Price': internet_price_list,
    'Trim': trim_list,
    'Transmission': transmission_list,
    'Class': class_list
})

# Close the WebDriver
driver.quit()
```
